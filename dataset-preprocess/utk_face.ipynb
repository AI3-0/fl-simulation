```python
def age_group(age):
    if age >=0 and age < 18:
        return 1
    elif age < 30:
        return 2
    elif age < 80:
        return 3
    else:
        return 4

dir = '../data/UTKFace/'
image_ids = os.listdir(dir)
labels = [] # age
protected_variables = []    # gender*10  + race
ages = []
genders = []
races = []
for image in image_ids:
    try:
        age, gender, race, _ = image.split('_')
        age, gender, race = age_group(int(age)), int(gender), int(race)
        ages.append(age)
        genders.append(gender)
        races.append(race)
    except ValueError:
        print(image)

ages = torch.tensor(ages, dtype=torch.long)
genders = torch.tensor(genders, dtype=torch.long)
races = torch.tensor(races, dtype=torch.long)

# labels = torch.tensor(labels, dtype=torch.long)
# protected_variables = torch.tensor(protected_variables, dtype=torch.long)
```
