```python
import matplotlib.pyplot as plt
import pickle


with open("../results/compas/R[500]F[0]LR[0.001]D[noniid]C[40]B[0.5].pkl", 'rb') as f:
    baseline = pickle.load(f)
with open("../results/compas/R[500]F[1]LR[0.001]D[noniid]C[40]B[0.5].pkl", 'rb') as f:
    feddyn = pickle.load(f)

x = [(i+1) * 5 for i in range(len(baseline['acc']))]

plt.figure(dpi=120)
plt.grid(b=True, which='major', axis='both')
plt.title('acc & auc_diff')
acc = baseline['acc']
auc_diff = [e['diff'] for e in baseline['auc']]
plt.plot(x, acc, label = 'accuracy_baseline')
plt.plot(x, auc_diff, label='fairness_baseline')

acc = feddyn['acc']
auc_diff = [e['diff'] for e in feddyn['auc']]
plt.plot(x, acc, label = 'accuracy_feddyn')
plt.plot(x, auc_diff, label='fairness_feddyn')

plt.xlabel("Round") 
plt.ylim(0.0, 0.7)
plt.legend(loc='lower right')
plt.show()
```
