```python
from utils import moving_average
import matplotlib.pyplot as plt
import pickle

with open("../results/celeba/R[50]F[0]LR[0.1]D[iid]C[100]B[0.3]FR[0.2]_E[5]_LD[0.998].pkl", 'rb') as f:
    iid = pickle.load(f)
with open("../results/celeba/R[50]F[0]LR[0.1]D[seperate]C[100]B[0.3]FR[0.2]_E[5]_LD[0.998].pkl", 'rb') as f:
    seperate = pickle.load(f)

x = [(i+1) * 5 for i in range(len(seperate['acc']))]

fig, ax1 = plt.subplots(figsize=(10, 8))
ax1.set_title('CelebA iid vs seperate (fedavg)')

ax1.set_xlabel("Round")
ax1.set_ylabel('acc', color = 'red')

plt.grid(b=True, which='major', axis='both')
ax1.plot(x, moving_average(iid['acc']), label = 'acc iid', color='orange')
ax1.plot(x, moving_average(seperate['acc']), label = 'acc seperate', color='red')
ax1.legend(loc='upper right')
ax1.tick_params(axis ='y', labelcolor = 'red') 
# ax1.set_ylim(0.9, 0.9175)
ax2 = ax1.twinx()

ax2.set_ylabel('fair', color='blue')
ax2.tick_params(axis ='y', labelcolor = 'blue') 
ax2.plot(x, moving_average(iid['fair']), label = 'Δacc iid', color='cyan')
ax2.plot(x, moving_average(seperate['fair']), label = 'Δacc seperate', color='blue')
ax2.legend(loc='lower right')
plt.show()
```
